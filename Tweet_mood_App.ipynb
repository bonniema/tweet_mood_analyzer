```python
#Load pre-trained model

from keras.models import model_from_json
model_json = open('mood_model.json','r').read()
model = model_from_json(model_json)
model.load_weights('mood_model.h5')
model.summary()
```

```python
#load the tokenizer
import pickle
tokenizer = pickle.load(open('mood_tokenizer.pickle','rb'))
tokenizer.word_index
```

```python
#now can load a tweet to test our loaded model

import numpy as np
from keras.preprocessing.sequence import pad_sequences
labels=['anger','disgust','fear','guilt','joy','sadness','shame']
tweet =['I am feeling on top of the world']

#turn it into numbers and then pad them
encoded_tweet = tokenizer.texts_to_sequences(tweet)
padded_tweet = pad_sequences(encoded_tweet, maxlen=256)
```

```python
#The probablity of mood
model.predict_proba(padded_tweet)
```

```python
np.argmax(model.predict_proba(padded_tweet))
```

```python
for label, prob in zip(labels, model.predict_proba(padded_tweet)[0]):
    print(label, ":", prob)
```

The end goal of our app is to let people type a tweet and return the mood.
